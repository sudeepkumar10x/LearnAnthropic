```python
# load env variables from .env file
from dotenv import load_dotenv
load_dotenv()

# create an instance of the Anthropic client and specify the model to use
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

## Added Temperature as one of the parameter in Chat function

```python
# Define helper functions to read latest text/ response & append it into message to 
# send it to the model and print the response

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system=None, temperature=1.0):
    kwargs = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }
    
    if system:
        kwargs["system"] = system
    
    message = client.messages.create(**kwargs)
    
    return message.content[0].text
```

# Exercise

## Using Temperature as 0.0

```python
# Start with an empty message list
messages = []

# Add the initial user question
add_user_message(messages, "Generate a one sentence bolliwood movie idea")

# Without system prompt
answer_temp_0 = chat(messages, temperature=0.0)

print(answer_temp_0)
```

### Every time you run this same prompt, it will generate almmost same response.

## Using Temperature as 1.0

```python
# Start with an empty message list
messages = []

# Add the initial user question
add_user_message(messages, "Generate a one sentence bolliwood movie idea")

# Without system prompt
answer_temp_1 = chat(messages, temperature=1.0)

print(answer_temp_1)
```
